# Pryngles System Demo

```python
from pryngles import *
```

## Single ringed planet

```python
sys=System()
S=sys.add("Star")
P=sys.add("Planet",primary=S,nspangles=200,m=1e-3,radius=0.3,x=3,vy=np.sqrt(sys.sim.G/3))
R=sys.add("Ring",primary=P,fi=2,fe=3,i=40*Consts.deg,roll=0*Consts.deg)
sys.spangle_system()
```

```python
sys.sg._interact_plot2d()
```

## Two planets, one of them ringed

```python
nspangles=500

sys=System()

S=sys.add("Star",bhash="Star",nspangles=nspangles,m=1,radius=1,x=0,y=0,vy=0)
P=sys.add("Planet",primary=S,bhash="Planet",nspangles=nspangles,radius=0.2,m=1e-3,x=2,vy=np.sqrt(sys.sim.G/2))
M=sys.add("Planet",primary=P,bhash="Moon",nspangles=nspangles,radius=0.1,m=1e-6,x=0,y=-3,vx=np.sqrt(sys.sim.G/3),vy=0)
R=sys.add("Ring",primary=P,bhash="Ring",nspangles=nspangles,fi=1.3,fe=2.3,i=20*Consts.deg,roll=180*Consts.deg)

sys.spangle_system()
```

```python
sys.sg._interact_plot2d(center_at="Ring")
```

```python
Plot.animate_rebound(sys.sim,tend=5,nsnap=50)
```

```python
sys.animate_integration(tini=0,tend=10,nsnap=50,interval=300,coords="ecl")
```
